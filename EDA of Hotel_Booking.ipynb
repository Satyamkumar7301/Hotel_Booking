## Cancellations 

```python
iris.isna().sum()
```

### Cancellations per months

```python
iris_2 = iris.groupby(['hotel','arrival_date_month']).agg({'is_canceled':'sum', 'meal':'count'}).reset_index().rename(columns={'is_canceled':'cancelled_bookings', 'meal':'total_bookings'})
iris_2['cancellation_percentage'] = (iris_2['cancelled_bookings']/iris_2['total_bookings'])*100
iris_2.head()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='arrival_date_month', y='cancellation_percentage', hue='hotel', data=iris_2, order=months)
plt.ylim((0,100))
plt.xticks(rotation=30)
plt.xlabel("Months")
plt.ylabel("% Cancellation")
plt.title("Cancellations per month")
```

#### For City Hotels the percentage cancellations are similar across all months whereas for Resort Hotels there are more cancellations during mid-year months of the year. 

### Effect of Lead time on Cancellations 

```python
iris_2 = iris.groupby('lead_time').agg({'is_canceled':'sum', 'hotel':'count'}).reset_index().rename(columns={'is_canceled':'cancelled_bookings', 'hotel':'total_bookings'})
iris_2['cancellation_percentage'] = (iris_2['cancelled_bookings']/iris_2['total_bookings'])*100
iris_2.head()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='lead_time', y='cancellation_percentage', data=iris_2)
plt.ylim((0,100))
plt.xlabel("Lead Time (days)")
plt.ylabel("% Cancellation")
plt.title("Effect of Lead time on Cancellations")
```

#### Bookings having less lead time are less often cancelled whereas high lead time bookings are more often cancelled. It shows positive correlation. 

### Cancellations by type of Guests (Repeated/NonRepeated) 

```python
plt.figure(figsize=(8,5))
sns.countplot(x = "is_canceled", hue = 'is_repeated_guest', data = iris)
plt.legend(['New Guest', 'Repeated Guest'])
plt.xticks(ticks=[0,1], labels=['Not Cancelled', 'Cancelled'])
plt.title("Cancellations by repeated guests")
```

#### We observe that repeated guests cancels their bookings less often than new guests. 

```python
plt.figure(figsize = (12,8))
sns.countplot(x = "market_segment", hue = 'is_canceled', data = iris)
plt.xlabel("Market Segment")
plt.legend(['Not Cancelled', 'Cancelled'])
plt.title("Cancellation by Market Segments")
```

#### We can see that Group Bookings gets cancelled more often. Direct segment has quite low cancellation rate compared to total number of bookings. For Offline TA/TO & Online TA have almost 50% cancellation rates.
